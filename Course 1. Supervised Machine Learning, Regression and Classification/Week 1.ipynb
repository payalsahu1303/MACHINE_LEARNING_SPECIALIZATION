# Linear Regression using Gradient Descent

## Linear Regression with One Variable (w)
### Problem Statement 1:
You are given the following dataset representing the number of hours studied (`x`) and the corresponding test scores (`y`):  

| Hours Studied (`x`) | Test Score (`y`) |
|----------------|-----------|
| 1  | 50 |
| 2  | 55 |
| 3  | 65 |
| 4  | 70 |
| 5  | 80 |

Use **Gradient Descent** to find the optimal weight (`w`) in the hypothesis function: **f(x) = wx**

### Tasks:
1. Initialize `w = 0` and use **Mean Squared Error (MSE)** as the cost function:  
   $$ J(w) = \frac{1}{2m} \sum_{i=1}^{m} (wx_i - y_i)^2 $$
2. Update `w` using **Gradient Descent** with a learning rate `Î± = 0.01`:  
   $$ w := w - \alpha \cdot \frac{1}{m} \sum (wx_i - y_i)x_i $$
3. Run for **1000 iterations** and find the optimal `w`.


```python
import numpy as np
import matplotlib.pyplot as plt

# Dataset (Hours Studied, Test Score)
X = np.array([1, 2, 3, 4, 5])  # Hours Studied
y = np.array([50, 55, 65, 70, 80])  # Test Scores
m = len(X)  # Number of data points

# Initialize parameter
w = 0  
alpha = 0.01  
iterations = 1000  

# Gradient Descent
cost_history = []
for _ in range(iterations):
    predictions = w * X
    error = predictions - y
    cost = (1 / (2 * m)) * np.sum(error ** 2)
    cost_history.append(cost)
    gradient = (1 / m) * np.sum(error * X)
    w = w - alpha * gradient

# Plot Cost Function
plt.plot(cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Cost Reduction')
plt.show()

print(f"Optimal w: {w:.4f}")
```

## 2. Medium Level (Two Variables - w, b)
### Problem Statement 2:
You are given a dataset representing house size (`x`) and house price (`y`):

| House Size (`x`) (sq. ft) | Price (`y`) (in $1000s) |
|----------------|-----------|
| 800  | 150 |
| 1000 | 180 |
| 1200 | 200 |
| 1500 | 250 |
| 1800 | 300 |

Use **Gradient Descent** to find the optimal weight (`w`) and bias (`b`) in the hypothesis function:**$$f(x) = wx + b$$**

### Tasks:
1. Initialize `w = 0`, `b = 0` and use the **Mean Squared Error (MSE)** cost function:  

$$
J(w, b) = \frac{1}{2m} \sum_{i=1}^{m} (wx_i + b - y_i)^2
$$

2. Update `w` and `b` using **Gradient Descent** with a learning rate `alpha = 0.01`:  

$$
w := w - \alpha \cdot \frac{1}{m} \sum (wx_i + b - y_i)x_i
$$

$$
b := b - \alpha \cdot \frac{1}{m} \sum (wx_i + b - y_i)
$$

3. Run for **1000 iterations** and find the optimal `w, b`.


```python
import numpy as np
import matplotlib.pyplot as plt

X = np.array([800, 1000, 1200, 1500, 1800])  # House Size (sq.ft)
y = np.array([150, 180, 200, 250, 300])  # Price (in $1000s)
m = len(X)  # Number of data points

# Initialize parameters
w, b = 0, 0  
alpha = 0.0000001  # Small learning rate due to large X values
iterations = 1000  

# Gradient Descent
cost_history = []
for _ in range(iterations):
    predictions = w * X + b
    error = predictions - y
    cost = (1 / (2 * m)) * np.sum(error ** 2)
    cost_history.append(cost)
    
    # Update w and b
    w_gradient = (1 / m) * np.sum(error * X)
    b_gradient = (1 / m) * np.sum(error)
    w = w - alpha * w_gradient
    b = b - alpha * b_gradient

# Plot Cost Function
plt.plot(cost_history)
plt.xlabel('Iterations')
plt.ylabel('Cost')
plt.title('Gradient Descent Cost Reduction')
plt.show()

print(f"Optimal w: {w:.4f}, Optimal b: {b:.4f}")
```
